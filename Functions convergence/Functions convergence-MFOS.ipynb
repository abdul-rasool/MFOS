```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes
csv ='A:/.../F2..csv' 

my_df = pd.read_csv(csv, encoding = 'latin-1')
```

```python
fig, ax = plt.subplots(figsize=(8,7))
ax.set_facecolor('white')
ax.plot(my_df['MFO'], label="MFO",linewidth=2)
ax.plot(my_df['FFA'], label="FFA",linewidth=2)
ax.plot(my_df['GWO'], label="GWO",linewidth=2)
ax.plot(my_df['DE'], label="DE",linewidth=2)
ax.plot(my_df['MVO'], label="MVO",linewidth=2)
ax.plot(my_df['HHO'], label="HHO",linewidth=2)
ax.plot(my_df['MFOS'], label="MFOS",linewidth=2)
plt.rcParams.update({'font.sans-serif':'Times New Roman'})
ax.set_title('Convergence Curve', fontsize=34, weight='bold')
ax.set_xlabel('Number of Iterations', fontsize=26)
ax.set_ylabel('Best fitness scores', fontsize=26)
ax.legend(loc='upper right', prop={"size":20})
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.savefig('new_test.png',dpi=800, bbox_inches = 'tight',pad_inches = 0.1)
```
